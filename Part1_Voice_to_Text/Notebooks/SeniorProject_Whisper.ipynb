```python
import whisper

def voice_to_text(audio_file):
  model = whisper.load_model("base")

  print(f"Processing audio: {audio_file}")

  result = model.transcribe(audio_file)
  result = result["text"].lower()
  return result

# converted text
converted_text = []
i = 0
for audio in audios:
  audio_file = audio_folder + f"/sample-{i}.wav"
  text = voice_to_text(audio_file)
  i += 1
  converted_text.append(text)


# print the converted_text list
n = 0
for text in converted_text:
  print(f"Audio #{n}: {text}")
  print("-" * 20)
  n += 1
```

```python
# Function to remove commas and end points
def remove_punctuation(text):
  return "".join([char for char in text if char not in (",.?")])


# check accuracy
wrong_converted_text = []
wrong_converted_num = []
total_correct_converted = 0
num = 0
text_clean = ""
orginial_clean = ""

for text in converted_text:
  text = remove_punctuation(text).strip()
  original_clean = remove_punctuation(text_csv.loc[num, "text"]).strip()
  if text == original_clean:
    total_correct_converted += 1
  else:
    wrong_converted_num.append(num)
    wrong_converted_text.append(text)
  num += 1

print(f"Correctly converted text: {total_correct_converted}")
print(f"Total number of audios: {len(converted_text)}")
print(f"Wrongly converted text: {len(wrong_converted_text)}")

print("-" * 30)

if(len(wrong_converted_text) > 0):
  for wrong_text, wrong_num in zip(wrong_converted_text, wrong_converted_num):
    print(f"Audio # {wrong_num}: {wrong_text}")
```
